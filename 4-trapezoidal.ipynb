## Implementing trapezoidal rule
against
$$
\displaystyle \int_{-\pi/2}^{ \pi/2 } \frac{1}{\sqrt{1-\sin^2{x} / 2}} dx
$$

```python
def func(x):
    return 1 / (np.sqrt(1 - (np.sin(x)**2 /2)))

def trapezoid(start, end, n):
    _sum = 0
    
    _sum += func(start) / 2
    _sum += func(end) / 2
    
    for k in range(1, n-1):
        _sum += func(start + k * (end - start) / (n-1))
        
    _sum *= (end - start) / (n-1)
    
    return _sum
```

```python
results = []
ns = []

for cnt in range(1, 20):
    n = 2 ** cnt
    
    results.append(trapezoid(-np.pi / 2, np.pi / 2, n))
    ns.append(n)
    
results[-1]
```

## Visualization

plot value

```python
plt.plot(ns, results)
```

plot ratio

```python
ratioes = []
for cnt in range(1, len(results)):
    ratioes.append(results[cnt] / results[cnt-1])
    
plt.plot(ratioes)
```
